```python
import pyfastx
fa = pyfastx.Fastx('rosalind_gcc.gz')
for name,seq in fa:
    print(name)
    print(seq)
for seq in fa:
    print(seq.name)
    print(seq.seq)
#FASTA 模块，注意不是FASTX模块，FASTX模块是用来处理FASTQ文件的
#读取Fasta文件，并且支持随机访问其中的任意序列。
fa = pyfastx.Fasta('rosalind_gcc.gz')
print("这是",fa[0].name)
print(len(fa))
gc_max=0#注意这个两行不要写在for循环里面，否则每次循环都会被初始化为0
gc_maxseq=0
for i in range(len(fa)-1):
    if fa[i].gc_content>gc_max:
        gc_max=fa[i].gc_content
        gc_maxseq=fa[i].name
print(gc_maxseq)
print(gc_max)
#这个有一个很容易出的错误：生成pyfastx.Fasta('rosalind_gcc.gz')这一步的时候会默认添加一个新的以fxi为结尾的文件，如果新调入文件运行且名字重复没有把之前的fxi文件删掉的话就会一直调用之前的数据
```
