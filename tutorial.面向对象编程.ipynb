```python
#实例方法：
#语法：
#class 类名(继承列表):
#    def 实例方法名(self,形式参数1,形式参数2,...):
#        "文档字符串"
#        语句

#说明：
#实例方法是实质是函数，是定义在类内的函数
#实例方法属于类的属性
#实例方法的第一个参数代表调用这个实例方法的对象，一般命名为self
#实例方法如果没有return语句，返回None

#实例方法的调用语法：
#实例.实例方法名(调用参数)
#或者
#类名.实例方法名(实例,调用参数)

#eg：我的理解是类使得定义的函数的参数变得更加明确，不用再去考虑参数的顺序，只需要记住参数的意义即可
class Dog:
    def __init__(self,k,c,a):
        self.kinds=k
        self.color=c
        self.age=a
    def say(self):
        print('我是一只%s,我的颜色是%s'%(self.kinds,self.color))
    def eat(self,that):
        "喂狗进食时在food属性记住吃的是什么"
        print("小狗正在吃",that)
        self.food=that#注意这个顺序,先定义food属性，再赋值
    def food_info(self):
        print("小狗吃了",self.food)
    def run(self,speed):
        print("小狗正以",speed,"速度奔跑")#这里的speed是形参，不是实参
    def get_color(self):
        return self.color
dog1=Dog('哈士奇','黑白',1)
print(dog1.kinds,dog1.color,dog1.age)
 
dog1.say()#用对象调用实例方法
Dog.say(dog1)#用类调用实例方法

dog1.eat('骨头')#注意因为eat不是__init__,不能像__init__那样直接传参，而是要用调用的手段传实参
print(dog1.food)
dog1.food_info()
```

# 类方法 @classmethod
类方法是只能访问类变量的方法
类方法需要使用@classmethod装饰器定义
类方法的第一个参数是类的实例，约定写为cls
说明：
类实例和对象实例都可以调用类方法
类方法不能访问实例变量

```python
#eg：
class ICBC:
    moneys=100000000 
    @classmethod
    def total_money(cls):
        print("工商银行总资金数：",cls.moneys)#哦哦哦就是说mooneys是类变量，所以要用cls类方法调用

    def __init__(self,b):
        self.branch=b
        self.money=50000
        self.__class__.moneys-=50000#这里的moneys是类变量，所以要用类对象调用
    def address(self):
        print("分行在：",self.branch)
ICBC.total_money()#返回的是类变量
b1=ICBC('天坛支行')
b1.total_money()#返回的是实例变量
```

# 类方法和实例方法对比：
1.类方法能够访问类变量，不能访问实例变量

实例方法能访问类变量（__class__属性间接访问），也能访问实例变量

2.类方法可以用实例来调用，也可以用类来调用

实例方法必须在调用时传入实例

# 静态方法 @staticmethod
静态方法是普通的函数

静态方法定义在类的内部，只能凭借该类和实例调用

静态方法需要使用@staticmethod装饰器定义

静态方法与普通函数定义相同，不需要传入self实例参数和class类参数

说明：类实例和对象实例都可以调用静态方法

静态方法不能访问类变量和实例变量

```python
class A:
    @staticmethod
    def myadd(a,b):
        return a+b
print(A.myadd(100,200))
a=A()
print(a.myadd(10,20))#静态方法可以通过实例对象调用
#挺没用的，主要是让这个函数和类关联起来，不用再去记住函数的名字，只需要记住类的名字即可
```

# 实例方法，类方法，静态方法总结：
不想访问类变量和实例变量（属性），用静态方法

只想访问类内变量，不想访问实例属性用类方法

既想访问类内变量，也想访问实例变量用实例方法

# 特殊属性@property
用来模拟一个属性，通过@property装饰器可以对模拟属性赋值和取值加以控制，实现其它语言所拥有的getter和setter功能

```python
#eg:
class Circle:#圆类
    def __init__(self,r):
        self.radius=r #半径
    @property
    def area(self):#面积
        return 3.14*self.radius**2
    @area.setter
    def area(self,a):#a代表面积（沿用了上面area这个函数，多传了a这个面积参数）
        self.radius=(a/3.14)**0.5
        print("更改面积，半径相应更改为：",self.radius)

c1=Circle(10)
print(c1.area)
c1.area=100
```

# 类的__slots__属性
作用：1.限定一个类创建的实例只能有固定的实例属性，不允许对象添加列表以外的实例属性
2.防止用户因错写属性名称而发生程序错误

说明：__slots__属性是一个列表，列表的值是字符串
含有__slots__属性的类所创建的实例对象没有__dict__属性，即此实例不用字典来存储属性

```python
#eg:
class Student:
    __slots__=('name','age','score')
    def __init__(self,n,a,s):
        self.name=n
        self.age=a
        self.score=s
        #或者写成：self.name,self.age,self.score=n,a,s
s1=Student('小张',20,100)
print(s1.name,s1.age,s1.score)
s1.age=30
print(s1.score)
#s1.Scre=200 #报错，因为slots限制了属性，只能是name,age,score
```

# 异常（基础）
## 什么是错误
错误是由于逻辑或语法错误等，导致一个城西已无法正常执行的问题
## 什么是异常
异常是程序出错时标识的一种状态，当异常发生时，程序不会再向下执行，而转去调用此函数的地方，待处理相应的错误并恢复为正常状态

```python
#try语句(此语法有两种写法)
#try/except/else/finally
#语法：
try:
    可能触发异常的语句
except 错误类型1[as 变量1]：
    异常处理语句1
except 错误类型2[as 变量2]：
    异常处理语句2
except (错误类型3,错误类型4) [as 变量3]：
    异常处理语句3
……
except:
    异常处理语句other #发生异常警告，其他什么也不做
else:
    未发生异常的语句
finally:
    最终语句
```

```python
#本程序用来演示异常处理
def div_apple(n):
    print("现在有",n,"个苹果，你想分给几个人？")
    d=int(input("请输入人数："))
    print("每个人分了",n/d,"个苹果")

try:
    div_apple(10)
except ValueError as err:
    print("发生值错误")
    print("错误的原因是：",err)
except ZeroDivisionError:
    print("发生了被零除的错误")
except:
    print("程序出现错误，已纠正！")
else:
    print("没有发生异常")
#finally:
#    print("我是finally语句，我一定会被执行")
print("程序正常结束")
```

# 运算符重载
用自定义的规则实现实例之间的运算符操作或函数操作


大概就是让两个不同类型变量不用重复定义一个新函数

```python
#对象转字符串函数重载
repr(obj)#返回一个对象的字符串形式，供解释器读取
str(obj)#返回一个对象的字符串形式，供用户读取

#对象转字符串函数重载方法
#repr(obj)函数的重载方法：
   def__repr__(self):
        ...
#str(obj)函数的重载方法：
    def__str__(self):
          ...
```

```python
#eg:
class MyNumber:
    def __init__(self,v):
        self.data=v
    def __repr__(self):#当语法学一下，经常用于比如从网站上搞下来的数据类型无法直接使用,用于数据清洗
        return 'MyNumber('+repr(self.data)+')'
    def __str__(self):
        return '数字是('+str(self.data)+')'
n1=MyNumber(100)
n2=MyNumber(200)
print(repr(n1))
print(str(n2))
print(n2)#等同于print(str(n2))
print(n1)#等同于print(str(n1))
```

```python
#算数运算符重载：
+  __add__
-  __sub__
* __mul__
/ __truediv__
// __floordiv__
% __mod__
** __pow__

#二元运算符重载的格式：
def __运算符__(self,other):#默认self是式子左边，other是式子右边
    ...
#说明：二元运算符的重载方法的参数列表中只能有两个参数
```

```python
#eg：试图把加法的意思改成改成两个列表合并的意思
class Mylist:
    def __init__(self,a):
        self.data=a.copy()
    def infos(self):
        return "Mylist("+str(self.data)+")"
    def __str__(self):#这个语句调用的是infos函数
        return self.infos()
    def myadd(self,other):
        r=[]
        r.extend(self.data)
        if type(other) == int:
            r.append(other)
        elif type(other) == Mylist:#在添加可迭代对象是，append在添加后不改变添加项的类型，添加之前是什么类型，添加之后就是什么类型；而extend在添加后，会将添加项进行迭代，迭代的元素挨个添加到被添加的数组中：
            r.extend(other.data)
        else:
            raise ValueError("不支持的操作")
        return Mylist(r)
    def __add__(self,other):
        return self.myadd(other)#意思是这个调用的是上面函数的结果
L1=Mylist([1,2,3])
L2=Mylist([4,5,6])
L3=L1.myadd(L2)
print("L3=",L3)
         
```

```python
#重载说明：
#运算符重载方法的参数已经有固定的含义，不可改变原有意义
#除__call__方法意外，其他重载方法的参数个数不可改变
```

# 内建函数重载：
__abs__(self)：绝对值
__len__(self)：长度
__reversed__(self)：反转
__round__(self,n)：四舍五入

```python
class Mylist:
    def __init__(self,a):
        self.data=a.copy()
    def __str__(self):
        return "Mylist("+str(self.data)+")"
    def __abs__(self):
        L=[abs(x) for x in self.data]#这句好妙
        return Mylist(L)
L1=Mylist([1,-2,3,-4,5])
L2=abs(L1)
print(L2)
```

```python
#eg:
class even:
    "偶数类，用于表达偶数的有序列表"
    def __init__(self,start,end):
        self.begin=start
        self.end=end #不包含end
    def __contains__(self,e):
        if e<self.begin:
            return False
        if e>=self.end:
            return False
        if e%2==0:
            return True
        return False#省略了else
e1=even(0,10)
print(2 in e1)
```
